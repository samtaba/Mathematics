# Optimization

The optimization process involves declaring an objective function, the variables, and the constraints of the problem.

- The objective functon expresses the quantity we wannt to minimize or maximize. 
- The variables are the unknowns in the system, the values we want to find that will make the objective function a maximum or minimum.
- The constraints are the functions that describe the relationships among the variables and that define the allowable values for the variables.

Optimization algorithms can be categorized into several branches, the main ones are 

- [Convex Optimization](https://en.wikipedia.org/wiki/Convex_function)
- [Non-Convex Optimization](https://www.cs.ubc.ca/labs/lci/mlrg/slides/non_convex_optimization.pdf)
- [Optimization Under Uncertainty](https://www.wiwi.uni-siegen.de/technologiemanagement/forschung/ouu.html?lang=de)

## 1. Unconstrained Univariate Minimizer (minimize_scalar)

SciPy's Optimize contains the minimize_scalar method which uses [Brent's methods](https://en.wikipedia.org/wiki/Brent%27s_method) (also known as the Brentâ€“Dekker method) to find the minima (turning point) of a univariate function (a function that takes a scalar as input).



```python
import numpy as np
from scipy.optimize import minimize_scalar
import matplotlib.pyplot as plt

# plt.style.use('ggplot')
```

Example: Find the minima of the function $ f(x) = (x-2)(x+1)^{2} $

```python
# function
f = lambda x: (x - 2) * (x + 1)**2

x = np.linspace(-2.5, 2.5, 100)

plt.plot(x, f(x))

plt.axhline(0, color='gray')
plt.axvline(0, color='gray')

plt.title("Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
res = minimize_scalar(f, method='brent')

# the minimia is at (1, -4)
res
```

## 2. Constrained Univariate Minimizer (minimize_scalar)

The interval constraint allows the minimization to occur between two fixed points, specified using the bounds parameter in minimize_scalar()

```python
res = minimize_scalar(f, bounds=(-2.5, 0), method='bounded')

# the minima is now at (-2.5, -10.125)
res
```

## 3. Root Finding (root_scalar)

There are multiple different root finding algorithms that can be used with SciPy.Optimize.root_scalar(). Most of these algorithms require the endpoints of an interval in which a root is expected. brentq is the best choice, but newton (or halley, secant) may be used instead. This is especially the case if the function is defined on a subset of the complex plane, and the bracketing methods cannot be used.

These are the available root finding algorithms:-

- bisect
- brentq
- brenth
- ridder
- toms748
- secant
- newton
- halley

```python
from scipy.optimize import root_scalar
```

```python
# function
f = lambda x: (x - 2) * (x + 1)**2 +3

x = np.linspace(-2.5, 2.5, 100)

plt.plot(x, f(x))

plt.axhline(0, color='gray')
plt.axvline(0, color='gray')

plt.title("Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

brentq requires the bounds to be specified for each root. It stops calculation when it finds the first root

```python
root = root_scalar(f, bracket=[-5, 5], method='brentq')

# it stops on the first root at x = -1.88
root
```

To find all the roots, the bounds have to be specified

```python
root1 = root_scalar(f, bracket=[-3, -1], method='brentq')

root2 = root_scalar(f, bracket=[0, 1], method='brentq')

root3 = root_scalar(f, bracket=[1, 2], method='brentq')

print(root1)
print()
print(root2)
print()
print(root3)
```

## 4. SciPy minimize

The scipy minimize() function provides a common interface to unconstrained and constrained minimization algorithms for multivariate scalar functions.

It supports the following algorithms:-
- [Nelder-Mead](https://en.wikipedia.org/wiki/Nelder%E2%80%93Mead_method) : Nelder-Mead Simplex Method

- [Powell](https://en.wikipedia.org/wiki/Powell%27s_method) : Powell's Method

- [CG](https://en.wikipedia.org/wiki/Conjugate_gradient_method) : Conjugate Gradient Method

- [BFGS](https://en.wikipedia.org/wiki/Broyden%E2%80%93Fletcher%E2%80%93Goldfarb%E2%80%93Shanno_algorithm) : Broydenâ€“Fletcherâ€“Goldfarbâ€“Shanno algorithm

- [Newton-CG](https://math.stackexchange.com/questions/1819410/newton-conjugate-gradient-algorithm) : Newton Conjugate Gradient Algorithm

- [L-BFGS-B](https://en.wikipedia.org/wiki/Limited-memory_BFGS) : Limited-memory Broydenâ€“Fletcherâ€“Goldfarbâ€“Shanno algorithm

- [TNC](https://en.wikipedia.org/wiki/Truncated_Newton_method) : Truncated Newton method

- [COBYLA](https://docs.scipy.org/doc/scipy/reference/optimize.minimize-cobyla.html) : Constrained Optimization by Linear Approximation

- [SLSQP](https://www.sciencedirect.com/science/article/abs/pii/B9780323958790502149) : Sequential least squares programming

- [trust-constr](https://en.wikipedia.org/wiki/Trust_region) : Trust Region Constrained Algorithms

- [dogleg](https://en.wikipedia.org/wiki/Powell%27s_dog_leg_method) : Powell's dog leg method

- [trust-ncg](https://epubs.siam.org/doi/10.1137/19M130563X) : Newton conjugate gradient trust-region algorithm

- [trust-exact](https://docs.scipy.org/doc/scipy/reference/optimize.minimize-trustexact.html) : Nearly Exact Trust-Region Algorithm

- [trust-krylov](https://docs.scipy.org/doc/scipy/reference/optimize.minimize-trustkrylov.html) : Nearly Exact Trust-Region Algorithm

- [custom](https://docs.scipy.org/doc/scipy/tutorial/optimize.html#custom-minimizers) : Custom Minimizer

```python
from scipy.optimize import minimize
```

Consider the following function

$ f(x,y) = 4 + (x-2)^{2} + (yâˆ’3)^{2} $

```python
def f(x):
  return 4 + (x[0]-2)**2.0 + (x[1]-3)**2.0
```

```python
x, y = np.meshgrid(np.linspace(-10, 15, 100), np.linspace(-10, 15, 100))
z = f([x,y])

fig = plt.figure(figsize =(8, 8)) 
ax = plt.axes(projection ='3d') 

my_cmap = plt.get_cmap('plasma') 

surf = ax.plot_surface(x, y, z, cmap = 'Spectral_r', alpha = 0.7, linewidth = 0.3, edgecolor = 'black') 

# fig.colorbar(surf, ax = ax, shrink = 0.7, aspect = 7) 

# Axis labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')

# Title
plt.title("3D surface plot")

plt.show() 
```

## 5. Unconstrained Minimization of Multivariate Scalar Functions

### a) Nelder-Mead Algorithm

```python
x0 = np.array([1.3, 0.7])

res = minimize(f, x0, method='nelder-mead', options={'xatol': 1e-8, 'disp': True})

# minimum is at (2, 3, 4)
res
```

### b) Broyden-Fletcher-Goldfarb-Shanno algorithm (BFGS)  

- This algorithm uses the gradient of the objective function. 
- If the gradient is not given by the user, then it is estimated using differences

Frst example, derivative is not provided

```python
x0 = np.array([3, -3])

res = minimize(f, x0, method='BFGS', options={'disp': True})

# minimum is at (2, 3, 4)
res
```

Second example, uses the derivative of the function in the form of a Jacobian

$$
\mathbf{J}
=
\frac{d \mathbf{f}}{d \mathbf{x}}
=
\left[ \frac{\partial \mathbf{f}}{\partial x_1}
\cdots \frac{\partial \mathbf{f}}{\partial x_n} \right]
=
\begin{bmatrix}
\dfrac{\partial f_1}{\partial x_1} & \cdots &
\dfrac{\partial f_1}{\partial x_n} \\
\vdots & \ddots & \vdots \\
\dfrac{\partial f_m}{\partial x_1} & \cdots &
\dfrac{\partial f_m}{\partial x_n}
\end{bmatrix}
$$

for the function defined above, 

$ \dfrac{\partial{f}}{\partial{x}} = 2(xâˆ’2) $

$ \dfrac{\partial{f}}{\partial{y}} = 2(yâˆ’3) $

```python
def f_jacobian(x):
    return [2*(x[0]-2), 2*(x[1]-3)]
```

```python
x0 = np.array([3, -3])

res = minimize(f, x0, method='BFGS', jac=f_jacobian, options={'disp': True})

# minimum is at (2, 3, 4)
res
```

### c) Newton Conjugate Gradient Algorithm

- This algorithm uses a conjugate gradient algorithm to invert the local Hessian
- It is a line search method: it finds a direction of search minimizing a quadratic approximation of the function and then uses a line search algorithm to find the (nearly) optimal step size in that direction.

<img src="./images/hessian_matrix.png" width="240px" style="filter:invert(1)">

```python
def f_hessian(x):
    return [[2, 0], [0, 2]]
```

```python
res = minimize(f, x0, method='Newton-CG', jac=f_jacobian, hess=f_hessian, options={'xtol': 1e-8, 'disp': True})

# minimum is at (2, 3, 4)
res
```

### d) Trust-Region Newton-Conjugate-Gradient Algorithm

```python
res = minimize(f, x0, method='trust-ncg', jac=f_jacobian, hess=f_hessian, options={'gtol': 1e-8, 'disp': True})

# minimum is at (2, 3, 4)
res
```

### e) Trust-Region Truncated Generalized Lanczos

```python
res = minimize(f, x0, method='trust-krylov', jac=f_jacobian, hess=f_hessian, options={'gtol': 1e-8, 'disp': True})

res
```

### f) Trust-Region Nearly Exact Algorithm

```python
res = minimize(f, x0, method='trust-exact', jac=f_jacobian, hess=f_hessian, options={'gtol': 1e-8, 'disp': True})

res
```

## 6. Constrained Minimization of Multivariate Scalar Functions

- The minimize function provides algorithms for constrained minimization
- It requires the constraints to be defined using the Bounds method
- Some algorithms require the constraints to be defined as a sequence of objects LinearConstraint and NonlinearConstraint

### a) Trust-Region Constrained Algorithm 

Define bounds using the Bounds method in SciPy Optimize

```python
from scipy.optimize import Bounds, LinearConstraint
```

The bound constraints $ 0 \leq x \leq 1 $ and $ -0.5 \leq y \leq 2.0 $ 
 are defined using a Bounds object.

```python
bounds = Bounds([0, -0.5], [1.0, 2.0])
```

The constraints $ x+2y \leq 1 $ and $ 2x+y = 1 $ can be written in the linear constraint standard format

<img src="./images/linear_constraint.png" width="300px" style="filter:invert(1)">

```python
linear_constraint = LinearConstraint([[1, 2], [2, 1]], [-np.inf, 1], [1, 1])
```

```python
res = minimize(f, x0, method='trust-constr', jac=f_jacobian, hess=f_hessian, constraints=[linear_constraint, nonlinear_constraint], options={'verbose': 1}, bounds=bounds)
```
