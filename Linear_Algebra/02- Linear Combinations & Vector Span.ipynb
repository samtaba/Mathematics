## Vector Linear Combinations

Let $ \{ \textbf{v}_1, \ \textbf{v}_2, \ \cdots \, \textbf{v}_p \} $ be vectors in $ \mathbb{R}^n $. A vector $ \textbf{b} $ is said to be a linear combination of the vectors $ \{ \textbf{v}_1, \ \textbf{v}_2, \ \cdots, \ \textbf{v}_p \} $ if there exists scalars $ x_1, x_2, \cdots, x_p $ such that 

$ x_1 \textbf{v}_1 + x_2 \textbf{v}_2 + \cdots + x_p \textbf{v}_p = \textbf{b} $.

### Linear Combination Problem

Given vectors $ \textbf{v}_1, \cdots, \textbf{v}_p $ and $ \textbf{b} $. Is $ \textbf{b} $ a linear combination of $ \textbf{v}_1, \cdots, \textbf{v}_p $ ?

The problem of determining if a given vector $ \textbf{b} $ is a linear combination of the vectors $ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p $ is equivalent to solving the linear system of equations with augmented matrix

$ [ \textbf{A} \ \ \textbf{b} ] = [ \textbf{v}_1 \ \ \textbf{v}_2 \ \ \cdots \ \ \textbf{v}_p \ \ \textbf{b} ] $

This provides for three possibilities:

- If the linear system is inconsistent, then $ \textbf{b} $ is not a linear combination of $ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p $
  
  In this case there doesn't exist scalars $ x_1, x_2, \cdots, x_p $ such that $ x_1 \textbf{v}_1 + x_2 \textbf{v}_2 + \cdots + x_p \textbf{v}_p = \textbf{b} $

- If the linear system is consistent and the solution is unique, then $ \textbf{b} $ can be written as a linear combination of $ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p $ in only one way

- If the linear system is consistent and the solution has free parameters, then $ \textbf{b} $ can be written as a linear combination of $ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p $ in infinitely many ways

### Example 1

Is the vector $ \textbf{b} $ a linear combination of vectors $ \textbf{v}_1 $ and $ \textbf{v}_2 $ ?

$ 
\textbf{b} = 
\begin{bmatrix}
7 \\
4 \\
-3 
\end{bmatrix}
\ \ \ \
\textbf{v}_1 =
\begin{bmatrix}
1 \\
-2 \\
-3
\end{bmatrix}
\ \ \ \
\textbf{v}_2 =
\begin{bmatrix}
\ 2 \ \\
\ 5 \ \\
\ 6 \
\end{bmatrix}
$

Solution:

We need to find $ x_1 $ and $ x_2 $ such that 

$ \textbf{b} = x_1 \textbf{v}_1 + x_2 \textbf{v}_2 $

Start by forming the augmented matrix $ [\textbf{v}_1\ \ \  \textbf{v}_2 \ \ \ \textbf{b}] $ then finding its RREF form

$ 
[\textbf{v}_1 \ \ \ \textbf{v}_2 \ \ \ \textbf{b}] 
     =
\begin{bmatrix}
1 & 2 & | & 7\\
-2 & 5 & | & 4\\
-5 & 6 & | & -3
\end{bmatrix}
$

```python
import numpy as np
from sympy import Matrix

A = np.array([[1, 2, 7], [-2, 5, 4], [-5, 6, -3]]);

# Convert to sympy matrix
Mat = Matrix(A)

# Get RREF (Reduced Row Echelon Form)
RREF = Mat.rref()[0]

print(np.array(RREF))
```

\end{bmatrix}
$

## Span

Let $ \{ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p \} \in \mathbb{R}^n $ be vectors. The set of all vectors that are a linear combination of $ \{ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p \} $ is called the span of $ \{ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p \} $, and is denoted as 

$ S = span\{\textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p\} $

If $ \textbf{b} \in span\{\textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p\} $ then $ \textbf{b} $ is a linear combination of $ \{ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p \} $. This implies there are scalars $ x_1, x_2, ..., x_n $ such that 

$ \textbf{b} = x_1 \textbf{v}_1 + x_2 \textbf{v}_2 + ··· + x_p \textbf{v}_p $

## Linear Independence

- The set of vectors $ \{\textbf{v}_1, \textbf{v}_2, ... , \textbf{v}_n\} $ is linearly independent if and only if $ \textbf{0} $ can be written in only one way as a linear combination of $ \{\textbf{v}_1, \textbf{v}_2, ..., \textbf{v}_n\} $. In other words, if 
  
  $ t_1\textbf{v}_1 + t_2 \textbf{v}_2 + ···+ t_n \textbf{v}_n = \textbf{0} $ 
  
  then necessarily the coefficients $ t_1, t_2, ..., t_n $ are all zero.
  
- A set of vectors $ \{\textbf{v}_1, \textbf{v}_2, ..., \textbf{v}_n\} $ is said to be linearly dependent if some $ \textbf{v}_j $ can be written as a linear combination of the other vectors, that is, if $ \textbf{v}_j \in span\{ \textbf{v}_1, ..., \textbf{v}_{j−1}, \textbf{v}_{j+1}, ..., \textbf{v}_n \} $. 

  If $ \textbf{v}_1, \textbf{v}_2, ..., \textbf{v}_n $ is not linearly dependent then we say that $ \textbf{v}_1, \textbf{v}_2, ..., \textbf{v}_n $ is linearly independent.

- The set $ \{\textbf{v}_1, \textbf{v}_2, ..., \textbf{v}_n\} $ is linearly independent if and only if the the rank of $ \textbf{A} $ is $ r = n $, that is, if the number of leading entries $ r $ in the REF (or RREF) of $ \textbf{A} $ is exactly $ n $.

### The maximum size of a linearly independent set

Let $ \textbf{v}_1, \textbf{v}_2, ..., \textbf{v}_p $ be a set of vectors in $ \mathbb{R}^n $. If $ p > n $ then $ \textbf{v}_1, \textbf{v}_2, ..., \textbf{v}_p $ are linearly dependent. Equivalently, if the vectors $ \textbf{v}_1, \textbf{v}_2, ..., \textbf{v}_p $ in $ \mathbb{R}^n $ are linearly independent then $ p ≤ n $.

## Bases Vectors

Let $ W $ be a subspace of a vector space $ V $. 

A set of vectors $ B = \{\textbf{v}_1, \cdots , \textbf{v}_k\} \in W $ is said to be a basis for $ W $ if 

- The set $ B $ spans all of $ W $, that is, $ W = span\{\textbf{v}_1, \cdots, \textbf{v}_k\} $, and
  
- The set $ B $ is linearly independent.

## Dimension Vector Space

Let $ V $ be a vector space. Then all bases of $ V $ have the same number of vectors.

Let $ V $ be a vector space. The dimension of $ V $, denoted $ dim V $, is the number of vectors in any basis of $ V $. The dimension of the trivial vector space $ V = {0} $ is
defined to be zero.

Let $ B = \{\textbf{v}_1, \cdots , \textbf{v}_n\} $ be vectors in $ \mathbb{R}^n $. If $ \textbf{B} $ is linearly independent then $ \textbf{B} $ is a basis for $ \mathbb{R}^n $. 
Or if $ span\{\textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_n\} = \mathbb{R}^n $ then $ \textbf{B} $ is a basis for $ \mathbb{R}^n $.

## Coordinates in an Orthonormal Basis

Let $ \textbf{B} = \{\textbf{u}_1, \textbf{u}_2, \cdots, \textbf{u}_n \} $ be an orthonormal basis for $ \mathbb{R}^n $. The coordinate vector of $ \textbf{x} $ in the basis $ \textbf{B} $ is

  $
  \textbf{x}_B =
  \begin{bmatrix}
    \textbf{u}_1 \cdot \textbf{x} \\
    \textbf{u}_2 \cdot \textbf{x} \\
    \vdots \\
    \textbf{u}_n \cdot \textbf{x}
  \end{bmatrix}
  $

## Linear Maps On Vector Spaces

Let $ T : V → U $ be a mapping of vector spaces. Then $ T $ is called a linear mapping if

- for any $ \textbf{u} $, $ \textbf{v} $ in $ V $ it holds that $ T(\textbf{u} + \textbf{v}) = T(\textbf{u}) + T(\textbf{v}) $, and

- for any scalar α and $ \textbf{u} $ in V it holds that $ T(α\textbf{v}) = αT(\textbf{v}) $.

Let $ T : V → U $ be a linear mapping.

- The kernel of $ T $ is the set of vectors $ \textbf{v} $ in the domain $ V $ that get mapped to the zero vector, that is, $ T(\textbf{v}) = 0 $. We denote the kernel of $ T $ by 
$ ker(T): ker(T) = {\textbf{v} ∈ V \ | \ T(\textbf{v}) = 0} $.

- The range of $ T $ is the set of vectors $ b $ in the co-domain $ U $ for which there exists at least one $ \textbf{v} $ in $ V $ such that $ T(\textbf{v}) = \textbf{b} $. We denote the range of $ T $ by 

  $ Range(T): Range(T) = {b ∈ U | \text{ there exists some } v ∈ U \text{ such that } T(v) = b} $.


Let $ T : V → U $ be a linear mapping. Then $ ker(T) $ is a subspace of $ V $ and $ Range(T) $ is a subspace of $ U $.
