## Gram-Schmidt Orthogonalization

Every nonzero finite-dimensional inner product space has an othonormal basis.

Gram-Schmidt Process: To convert a basis $ \{\textbf{u}_1, \textbf{u}_2, \cdots , \textbf{u}_r \} $ into an orthogonal basis $ \{\textbf{v}_1, \textbf{v}_2, \cdots , \textbf{v}_r \} $ perform the following r computations

step 1: $ \ \ \textbf{v}_1 = \textbf{u}_1 $

step 2: $ \ \ \textbf{v}_2 = \textbf{u}_2 - \dfrac{\textbf{u}_2 \cdot \textbf{v}_1}{|\textbf{v}_1|^2} \textbf{v}_1 $

step 3: $ \ \ \textbf{v}_3 = \textbf{u}_3 - \dfrac{\textbf{u}_3 \cdot \textbf{v}_1}{|\textbf{v}_1|^2} \textbf{v}_1  - \dfrac{\textbf{u}_3 \cdot \textbf{v}_2}{|\textbf{v}_2|^2} \textbf{v}_2 $

step 4: $ \ \ \textbf{v}_4 = \textbf{u}_4 - \dfrac{\textbf{u}_4 \cdot \textbf{v}_1}{|\textbf{v}_1|^2} \textbf{v}_1  - \dfrac{\textbf{u}_4 \cdot \textbf{v}_2}{|\textbf{v}_2|^2} \textbf{v}_2 - \dfrac{\textbf{u}_4 \cdot \textbf{v}_3}{|\textbf{v}_3|^2} \textbf{v}_3 $

$ \vdots $

step $ r $: $ \ \ \textbf{v}_r = \textbf{u}_r - \dfrac{\textbf{u}_r \cdot \textbf{v}_1}{|\textbf{v}_1|^2} \textbf{v}_1  - \dfrac{\textbf{u}_r \cdot \textbf{v}_2}{|\textbf{v}_2|^2} \textbf{v}_2 - \dfrac{\textbf{u}_r \cdot \textbf{v}_3}{|\textbf{v}_3|^2} \textbf{v}_3 - \cdots - \dfrac{\textbf{u}_r \cdot \textbf{v}_{r-1}}{|\textbf{v}_{r-1}|^2} \textbf{v}_{r-1} $

### Example

Let $ \mathbb{R}^3 $ have the Euclidean inner product. Use the Gram -Schmidt proces to transform $ \{\textbf{u}_1, \textbf{u}_2, \textbf{u}_3 \} $ into an orthonormal basis.

$
\textbf{u}_1 =
\begin{bmatrix}
1 \\
1 \\
0 
\end{bmatrix}
\ , \ \ \textbf{u}_2 =
\begin{bmatrix}
1 \\
2 \\
0
\end{bmatrix}
\ , \ \ \textbf{u}_3 = 
\begin{bmatrix}
0 \\
1 \\
2
\end{bmatrix}
$

**Solution:**

The vectors $ \{\textbf{u}_1, \textbf{u}_2, \textbf{u}_3 \} $ are neither orthogonal nor orthonormal. We will use the Gram-Schmidt method to convert them to a set of new orthogonal basis vectors $ \{\textbf{v}_1, \textbf{v}_2, \textbf{v}_3 \} $, subsequently we will normalize them to obtain the orthonormal vectors.

$ \textbf{u}_1 \cdot \textbf{u}_2 = 3 $

$ \textbf{u}_1 \cdot \textbf{u}_3 = 1 $

$ \textbf{u}_2 \cdot \textbf{u}_3 = 2 $

None of them are zero so not orthogonal.

Using Gram-Schmidt method, start creating the vectors

$ \textbf{v}_1 = \textbf{u}_1 = (1,1,0) $

$ \textbf{v}_2 = \textbf{u}_2 - \dfrac{\textbf{u}_2 \cdot \textbf{v}_1}{|\textbf{v}_1|^2} \textbf{v}_1 = (1,2,0)-\frac{3}{2} (1,1,0) = (-\frac{1}{2}, \frac{1}{2}, 0) = (-1,1,0) $

$ \textbf{v}_3 = \textbf{u}_3 - \dfrac{\textbf{u}_3 \cdot \textbf{v}_1}{|\textbf{v}_1|^2} \textbf{v}_1 - \dfrac{\textbf{u}_3 \cdot \textbf{v}_2}{|\textbf{v}_2|^2} \textbf{v}_2 = (0,1,2)-\frac{1}{2} (1,1,0) - \frac{1}{2} (-1,1,0) = (0,0,2)  $

Check these basis vectors are orthogonal

$ \textbf{v}_1 \cdot \textbf{v}_2 = (1,1,0).(-1,1,0) = 0 $

$ \textbf{v}_1 \cdot \textbf{v}_3 = (1,1,0).(0,0,2) = 0  $

$ \textbf{v}_2 \cdot \textbf{v}_3 = (-1,1,0).(0,0,2) = 0  $

Which are orthogonal. 

To make these vectors orthonormal

$ \textbf{v}_1 = (\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}},0) $

$ \textbf{v}_2 = (\frac{-1}{\sqrt{2}}, \frac{1}{\sqrt{2}},0) $

$ \textbf{v}_3 = (0, 0,1) $
