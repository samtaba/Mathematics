## Span

Let $ \{ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p \} \in \mathbb{R}^n $ be vectors. The set of all vectors that are a linear combination of $ \{ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p \} $ is called the span of $ \{ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p \} $, and is denoted as 

$ S = span\{\textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p\} $

If $ \textbf{b} \in span\{\textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p\} $ then $ \textbf{b} $ is a linear combination of $ \{ \textbf{v}_1, \textbf{v}_2, \cdots, \textbf{v}_p \} $. This implies there are scalars $ x_1, x_2, ..., x_n $ such that 

$ \textbf{b} = x_1 \textbf{v}_1 + x_2 \textbf{v}_2 + ··· + x_p \textbf{v}_p $

### Example 1

Given the vectors $ \textbf{v}_1 $, $ \textbf{v}_2 $ and $ \textbf{b} $ shown below. 

Is the vector $ \textbf{b} $ in the span of the vectors $ \textbf{v}_1 $ and $ \textbf{v}_2 $, or $ \textbf{b} \in span\{\textbf{v}_1, \textbf{v}_2\} $

$ 
\textbf{b} = 
\begin{bmatrix}
7 \\
4 \\
-3 
\end{bmatrix}
\ \ \ \
\textbf{v}_1 =
\begin{bmatrix}
1 \\
-2 \\
-5
\end{bmatrix}
\ \ \ \
\textbf{v}_2 =
\begin{bmatrix}
\ 2 \ \\
\ 5 \ \\
\ 6 \
\end{bmatrix}
$

Solution:

$ \textbf{b} \in span\{\textbf{v}_1, \textbf{v}_2\} $ if $ \textbf{b} $ can be expressed as a linear combination of $\{\textbf{v}_1, \textbf{v}_2\} $ i.e. there exists scalars $ x_1 $ and $ x_2 $ such that

$ x_1 \textbf{v}_1 + x_2 \textbf{v}_2 = \textbf{b} $

We have already evaluated this in Example 1 of Linear Combinations section

$
\textbf{b} = 3
\begin{bmatrix}
1 \\
-2 \\
-3
\end{bmatrix}
    + 2
\begin{bmatrix}
2 \\
5 \\
6
\end{bmatrix}
    =
\begin{bmatrix}
7 \\
4 \\
-3 
\end{bmatrix}
$

Therefore $ \textbf{b} \in span\{\textbf{v}_1, \textbf{v}_2\} $

$ \textbf{b} = x_1 \textbf{v}_1 + x_2 \textbf{v}_2 + x_3 \textbf{v}_3 $

### Example 3

The vectors $ \textbf{u} $, $ \textbf{v} $ and $ \textbf{w} $ are defined below. Show that $ \textbf{w} $ is in $ span\{ \textbf{u}, \textbf{v} \} $ 

$ 
\textbf{u} = 
\begin{bmatrix}
1 \\
1 \\
0 
\end{bmatrix}
,\ \ \ \
\textbf{v} =
\begin{bmatrix}
3 \\
2 \\
0
\end{bmatrix}
,\ \ \ \ 
\textbf{w} = 
\begin{bmatrix}
4 \\
5 \\
0 
\end{bmatrix}
$

Solution: 

For a vector to be in $ span\{ \textbf{u}, \textbf{v} \} $, it must be a linear combination of these vectors.

If $ \textbf{w} \in span\{ \textbf{u}, \textbf{v} \} $, we must be able to find scalars $ a $ and $ b $ such that 

$ \textbf{w} = a \textbf{u} + b \textbf{v} $ 

Writing this matrix form

$ 
\begin{bmatrix}
4 \\
5 \\
0 
\end{bmatrix}
    =
a
\begin{bmatrix}
1 \\
1 \\
0
\end{bmatrix}
    b 
\begin{bmatrix}
3 \\
2 \\
0 
\end{bmatrix}
$

To solve, write this as a system of linear equations

$
\begin{bmatrix}
1 & 3 \\
1 & 2 
\end{bmatrix}
\begin{bmatrix}
a \\
b 
\end{bmatrix}
    = 
\begin{bmatrix}
4 \\
5  
\end{bmatrix}
$

To solve this we can express the matrix in RREF or calculate the inverse matrix. 

```python
import numpy as np
import scipy as sp
from sympy import Matrix


# create the matrix
A = np.array([[1, 3, 4], [1, 2, 5]]);

# Convert to sympy matrix
A_Mat = Matrix(A)

# find RREF (Reduced Row Echelon Form)
RREF = A_Mat.rref()[0]

print(np.array(RREF), '\n')
```

The RREF results means $ a = 7 $ and $ b = -1 $

Therefore, 

$ \textbf{w} = 7 \textbf{u} - \textbf{v} $

We can also evaluate the inverse matrix

$
\begin{bmatrix}
a \\
b 
\end{bmatrix}
    =
\begin{bmatrix}
1 & 3 \\
1 & 2 
\end{bmatrix}^{-1} 
\begin{bmatrix}
4 \\
5  
\end{bmatrix}
$

```python
import numpy as np
import scipy as sp


# create the matrix
A = np.array([[1, 3,], [1, 2]]);


# find inverse using scipy
print(sp.linalg.inv(A))
```

$
\begin{bmatrix}
a \\
b 
\end{bmatrix}
    =
\begin{bmatrix}
-2 & 3 \\
1 & -1 
\end{bmatrix} 
\begin{bmatrix}
4 \\
5  
\end{bmatrix}
$

This means

$ a = -8 + 15 = 7 $ and $ b = 4 - 5 = -1 $ 

with the same result

$ \textbf{w} = 7 \textbf{u} - \textbf{v} $
