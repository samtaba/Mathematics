# Newton's Method

Newton's method can be used to find the root of a function, or the minima of a function. The algorithm varies a little between these two algorithms.

## Newton's Method for Root Finding

In this algorithm, we use the gradient of the function to get closer and closer to the root of the function.

The tangent cuts the x-axis at $ x_1 $, which will be a better approximation of the root. Now, drawing another tangent at $ (x_1, f(x_1)) $, which cuts the x-axis at $ x_2 $, which is a still better approximation and the process can be continued until the desired accuracy is achieved.

<img src="./images/newton-raphson-method.png" width="300px">

- Start with a close guess to the root, $ x_0 $ for example

  The equation of the tangent at that point is
  
  $ f(x) = mx + c \ \ \ $ 
  
- evaluate the gradient of the tangent at $ x_0 $
  
  $ m = f'(x_0) \ \ \ $ (by definition)

  $ \Rightarrow f(x_0) = f'(x_0) x_0 + c_0 $

- Now evaluate the value of the intercept $ c $

  $ c_0 = f(x_0) -  f'(x_0) x_0 $ 

- The equation of the tangent is therefore
  
  $ f(x) = f'(x_0) x + f(x_0) -  f'(x_0) x_0 $

- Now use $ x_1 $, the intercept of the tangent line with the x-axis, as the next point to iterate from.

  Using the equation of the tangent above and evaluating it at $ f(x_1) = 0 $
  
  $ f(x_1) = f'(x_0) x_1 + f(x_0) -  f'(x_0) x_0 = 0 $

  $ \Rightarrow f'(x_0) x_1 = f'(x_0) x_0 - f(x_0) $
  
  $ \Rightarrow x_1 = x_0 - \dfrac{f(x_0)}{f'(x_0)} $

- If we repeated this process and evaluate the tangent of the function at $ x_1 $, we would get closer and closer to the root of the equation. We can therefore write the iterative Newton's method as

  $ \boxed{x_{n+1} = x_n - \dfrac{f(x_n)}{f'(x_n)}} $ 

### Example

Evaluate the roots of the function $ f(x) = (x - 2) (x + 1)^2 +3 $ using Newton's method.

First, let's plot the function:

```python
import numpy as np
import matplotlib.pyplot as plt

# function
f = lambda x: (x - 2) * (x + 1)**2 + 3

x = np.linspace(-2.5, 2.5, 100)

plt.plot(x, f(x))

plt.axhline(0, color='gray')
plt.axvline(0, color='gray')

plt.title("Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

Now evaluate the first root starting at $ x = 2 $



```python
def newtons_method(start_point, function, der_function, iterations):
  x = start_point
  for i in range(0, iterations):
    x = x - function(x)/der_function(x)
    
  return x

f = lambda x: (x - 2) * (x + 1)**2 + 3
d_f = lambda x: 3*(x**2 -1)

start_point1 = 2
root1 = newton_method(start_point1, f, d_f, 50)
print(root1)  

start_point2 = 0
root2 = newton_method(start_point2, f, d_f, 50)
print(root2)  

start_point3 = -1.5
root3 = newton_method(start_point3, f, d_f, 50)
print(root3)  
```

## Newton's Method for Minima

To find the minima of a function, we need to find the roots of the derivative function, i.e. Newton's method needs to be applied to the derivative of the function rather than the function itself.

The minima can be obtained with the following iteration

$ \boxed{x_{n+1} = x_n - \dfrac{f'(x_n)}{f''(x_n)}} $ 

```python
import numpy as np
import matplotlib.pyplot as plt

# function
f = lambda x: (x - 2) * (x + 1)**2 + 3
d_f = lambda x: 3*(x**2 -1)

x = np.linspace(-2, 2, 100)

plt.plot(x, f(x), label='function')
plt.plot(x, d_f(x), label='derivative of function')

plt.axhline(0, color='gray')
plt.axvline(0, color='gray')

plt.title("Function and its derivative")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

### Example

Find the minima of the function $ f(x) = (x - 2) (x + 1)^2 + 3 $ using Newton's method.

This is done by evaluating the root of the derivative of this function.

```python
def newtons_method(start_point, function, der_function, iterations):
  x = start_point
  for i in range(0, iterations):
    x = x - function(x)/der_function(x)
    
  return x

d_f = lambda x: 3*(x**2 -1)
dd_f = lambda x: 6*x

start_point1 = 2
minima_1 = newtons_method(start_point1, d_f, dd_f, 50)
print(minima_1)  

start_point2 = -2
minima_2 = newtons_method(start_point2, d_f, dd_f, 50)
print(minima_2)  
```
