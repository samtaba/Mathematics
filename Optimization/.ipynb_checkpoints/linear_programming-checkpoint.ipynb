# Linear Programming

Linear programming, also called linear optimization, is a method to achieve the optimum outcome in a mathematical model whose requirements and objective are represented by linear relationships.

Linear programming problems take the form:

Find a vector $ \ \textbf{x} $ 

That maximizes $ \ \textbf{c}^T \textbf{x} $ 

Subject to $ \ \textbf{Ax} \le \textbf{b} $

and $ \ \textbf{x} \ge \textbf{0} $

### Example

Find the maximum solution to the following function:-

$ f(x,y) = 4x + 3y $

subject to the following constraints:

$ x ≥ 0 $  

$ y ≥ 2 $  

$ 2y ≤ 25 - x $  

$ 4y ≥ 2x - 8 $ 

$ y ≤ 2x - 5 $  

**Solution:**

The function $ f(x,y) $ is our objective function, $ x $ and $ y $ in this equation are constrained variables.

Let's graph this problem

```python
import numpy as np
import matplotlib.pyplot as plt

# first, plot the constraints region
# x > 0
x = np.linspace(0, 20, 2000)

# y >= 2
y1 = (x*0) + 2

# 2y <= 25 - x
y2 = (25-x)/2.0

# 4y >= 2x - 8 
y3 = (2*x-8)/4.0

# y <= 2x - 5 
y4 = 2 * x -5

# Make plot
plt.plot(x, y1, label=r'$y\geq2$')
plt.plot(x, y2, label=r'$2y\leq25-x$')
plt.plot(x, y3, label=r'$4y\geq 2x - 8$')
plt.plot(x, y4, label=r'$y\leq 2x-5$')
plt.xlim((0, 16))
plt.ylim((0, 11))
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

# Fill feasible region
y5 = np.minimum(y2, y4)

y6 = np.maximum(y1, y3)

plt.fill_between(x, y5, y6, where=y5>y6, color='grey', alpha=0.5)

plt.title("Linear Programming Constraints")

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# now plot the cost function

def cost_function(xdata, ydata):
  return 4*xdata + 3*ydata

x = np.linspace(0, 20, 2000)
y = np.linspace(0, 20, 2000)
    
X, Y = np.meshgrid(x, y)

f = cost_function(X, Y)

fig = plt.figure(figsize =(8, 8)) 
ax = plt.axes(projection ='3d') 

surf = ax.plot_surface(X, Y, f, cmap = 'Spectral_r', alpha = 0.7, linewidth = 0.3, edgecolor = 'black') 
```

The objective function maxima lies somewhere in the grey area in the graph above. That is between the vertices of the grey region. 
Since the cost function in this case is a 2d plane that lies above the grey area, we can deduce that its maxima will either lie on a vertex or a corner.

Let's find the solutions for each corner to find our maximum.

**Corner 1:**

$ y ≥ 2 \ \ $ (blue line) and $ \ \ 4y ≥ 2x - 8 \ $ (green line)   

Solve these to linear equations to find the intersection point.

$ y = 2 $

$ 2 = (2x - 8) / 4 $  
    
$ \Rightarrow x = 8 $

$ \Rightarrow f = 38 $  


**Corner 2:**

$ 2y ≤ 25 - x \ \ \ $ and $ \ \ \ y ≤ 2x - 5 $

Solve the two equations to find the intersection
  
$ (25 - x) / 2 = (2x - 5) $ 

$ \Rightarrow x = 7 $ 

$ \Rightarrow y = 9 $

$ \Rightarrow f = 55 $  


**Corner 3:**

$ 2y ≤ 25 - x $  and  $ 4y ≥ 2x - 8 $  

$ (25 - x) / 2 = (2x - 8) / 4 $

$ \Rightarrow x = 14.5 $ 

$ \Rightarrow y = 5.25 $

$ \Rightarrow f = 73.75 $  


**Corner 4:**

$ y ≥ 2 $ and $ y ≤ 2x - 5 $

$ 2 = 2x-5 $  

$ \Rightarrow x = 3.5 $  

$ \Rightarrow y = 2 $  

$ \Rightarrow f = 20 $ 

Therefore we get a maxima for $ f = 73.75 $ at $ x = 14.5 $ and $ y = 5.25 $
