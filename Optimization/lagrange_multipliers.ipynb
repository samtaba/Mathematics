# Lagrange Multiplier and Constrained Optimization

Lagrange multiplier is a technique for finding the local maxima and minima of a function subject to equation constraints. The technique is used to convert a constrained optimization problem into an unconstrained problem. The relationship between the gradient of the function and gradients of the constraints leads to a reformulation of the original problem as a Lagrangian function or Lagrangian.

Consider a function we want to maximize given a constraint 

$ \text{maximize} \ \ f(\textbf{x}) $

$ \text{subject to} \ \ g(\textbf{x}) = c $

Can be written as a single Lagrangian function

$ \mathcal{L}(\textbf{x}, \lambda) = f(\textbf{x}) + \lambda \cdot (g(\textbf{x}) - c) $

Where $ \lambda $ is known as the Lagrange multiplier and may be either added or subtracted.

$ c $ is the constraint constant.

The Lagrangian is differentiated to find the maximum/minimum values of the function

$ \dfrac{\partial \mathcal{L}}{\partial \textbf{x}} = 0 $

$ \dfrac{\partial \mathcal{L}}{\partial \lambda} = 0 $

### Example 1

Maximize utility $ f(x,y) = x^{\frac{1}{3}}y^{\frac{2}{3}} \ $ 

subject to the constraint 

$ \ g(x,y) = 150 - 5x = 0 $

**Solution:**

We start by evaluating this numerically by observing the values of the function on the line $ y = 150 - 5x $

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 150, 500)
y = np.linspace(0, 150, 500)

# add the restriction into the function
z = (x**(1/3)) * (150-5*x)**(2/3)

# remove all nan
z = z[~np.isnan(z)]

zmax = np.max(z)

print("z = ", zmax)

# now find the values of x and y 
max_index = np.where(z==zmax)
max_x = x[max_index[0][0]]
print("x = ", max_x)
print("y = ", 150 - 5*max_x)
```

The maximum value of this function on the constraint line is $ \ \ x = 9.92, \ \ y = 100.40, \ \ z = 46.42 $.

Contour plot the objective function with the constraint line.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 150, 500)
y = np.linspace(0, 150, 500)

X, Y = np.meshgrid(x, y)

Z = (X**(1/3)) * Y**(2/3)

y1 = 150-5*x

levels = np.linspace(0, 100, 10)

plt.figure(figsize=(7, 5))

contour = plt.contourf(X, Y, Z)
# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('f(x,y)=x*y Objective function')
plt.contourf(X, Y, Z, levels=levels)
plt.colorbar(contour, label='Z-values')


plt.plot(x, y1, c='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('y=150-5*x Constraint function')
ax = plt.gca()
ax.set_xlim([0, 150])
ax.set_ylim([0, 150])

# the max z point on the line 
plt.scatter(10, 100, c='r')

# Displaying the plot
plt.show()
```

We want to find the maximum value of the function along the red line above.

Write the Lagrangian:

$ \mathcal{L}(x, y, \lambda) = x^{\frac{1}{3}}y^{\frac{2}{3}} + \lambda (150 - 5x -y) $

differentiate and set equal to zero

$ \dfrac{\partial \mathcal{L}}{\partial x} = \dfrac{1}{3} x^{\frac{-2}{3}} y^{\frac{2}{3}} -5 \lambda = 0 \ \ \ \Rightarrow \ \lambda = \dfrac{1}{15} x^{\frac{-2}{3}} y^{\frac{2}{3}} $

$ \dfrac{\partial \mathcal{L}}{\partial y} = \dfrac{2}{3} x^{\frac{1}{3}} y^{\frac{-1}{3}} -\lambda = 0 \ \ \ \Rightarrow \ \lambda = \dfrac{2}{3} x^{\frac{1}{3}} y^{\frac{-1}{3}} $

$ \dfrac{\partial \mathcal{L}}{\partial \lambda} = 150 - 5x -y = 0 $

Equate the first two equations

$ \dfrac{1}{15} x^{\frac{-2}{3}} y^{\frac{2}{3}} = \dfrac{2}{3} x^{\frac{1}{3}} y^{\frac{-1}{3}} \ \ \ \ \Rightarrow y = 10 x $

Substitute this into third equation

$ 150 -5x -10x = 0 $ 

$ \Rightarrow x = 10, \ \ y = 100, \ \ z = x^{\frac{1}{3}}y^{\frac{2}{3}} = 46.416 $

### Example 2

Maximize utility $ f(x,y) = xy \ \ $ 

subject to the constraint 

$ \ \ g(x,y) = 10x + 20y = 400 $

**Solution:**

We start by evaluating this numerically by observing the values of the function on the line $ y = 20 - \frac{1}{2}x $

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 150, 500)
y = np.linspace(0, 150, 500)

# add the restriction into the function
z = x*(20-0.5*x)

# remove all nan
z = z[~np.isnan(z)]

zmax = np.max(z)

print("z = ", zmax)

# now find the values of x and y 
max_index = np.where(z==zmax)
max_x = x[max_index[0][0]]
print("x = ", max_x)
print("y = ", 20 - 0.5*max_x)
```

The maximum value of this function on the constraint line is $\ \ x = 20.14, \ \ y = 9.93, \ \ z = 199.99 $

Contour plot the objective function with the constraint line.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 50, 500)
y = np.linspace(0, 30, 500)

X, Y = np.meshgrid(x, y)

Z = X*Y

y1 = 20-0.5*x

levels = np.linspace(0, 100, 10)

plt.figure(figsize=(7, 5))

contour = plt.contourf(X, Y, Z)
# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('f(x,y)=x*y Objective function')
plt.contourf(X, Y, Z, levels=levels)
plt.colorbar(contour, label='Z-values')


plt.plot(x, y1, c='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('y=20-0.5*x Constraint function')
ax = plt.gca()
ax.set_xlim([0, 50])
ax.set_ylim([0, 30])

# the max z point on the line 
plt.scatter(20, 10, c='r')

# Displaying the plot
plt.show()
```

We want to find the maximum value of the function along the red line above.

Write the Lagrangian:

$ \mathcal{L}(x, y, \lambda) = x y - \lambda (10x + 20y - 400) $

differentiate and set equal to zero

$ \dfrac{\partial \mathcal{L}}{\partial x} = y - 10\lambda = 0 \ \ \ \ \Rightarrow 2y - 20\lambda = 0$

$ \dfrac{\partial \mathcal{L}}{\partial y} = x - 20\lambda = 0 $

$ \dfrac{\partial \mathcal{L}}{\partial \lambda} = -(10x + 20 y -400) = 0 \ \ \ \ \Rightarrow 10x + 20 y = 400 $

Equate the first two equations

$ 2y - 20\lambda = x - 20\lambda \ \ \ \ \Rightarrow x = 2y $

Substitute this into third equation

$ 10(2y) + 20 y = 400 \ \ \ \ \Rightarrow 40y = 400 $

$ \Rightarrow x = 20, \ \ y = 10, \ \ z = xy = 200 $
