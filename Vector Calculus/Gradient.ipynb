# Gradient of Scalar Field

The gradient of the scalar field $ \phi(x,y,z) $ is given by:

$
 \vec{\nabla} \phi = \left( \frac{\partial \phi}{\partial x}, \frac{\partial \phi}{\partial y}, \frac{\partial \phi}{\partial z} \right)
$

We can also write it as an operator 

$
\vec{\nabla} = \dfrac{\partial}{\partial x} \ \bold{i} + \dfrac{\partial}{\partial y} \ \bold{j} + \dfrac{\partial}{\partial z} \ \bold{k} 
$

**Interpretation:** A vector in the plane that points to increasing values of the function. Or, a vector that points towards highest values of contour plot.

### Example 1

Plot contour and gradients of the function 

$ f(x,y) = x^2 + y^2 $

with gradient

$ \vec{\nabla} f(x,y) = \dfrac{\partial f}{\partial x} \bold{i} + \dfrac{\partial f}{\partial y} \bold{j} = 2 x \ \bold{i} + 2 y \ \bold{j} $


```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(np.linspace(-1.2, 1.2, 100), np.linspace(-1.2, 1.2, 100))
z = x**2 + y**2

fig = plt.figure(figsize =(7, 7)) 
ax = plt.axes(projection ='3d') 

my_cmap = plt.get_cmap('plasma') 

surf = ax.plot_surface(x, y, z, cmap = 'Spectral_r', alpha = 0.7, linewidth = 0.3, edgecolor = 'black') 

# Axis labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')

# Title
plt.title("3D surface plot")

plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt


plt.figure(figsize=(7, 7))

wy = np.linspace(-1.2, 1.2, 100)

wx = np.linspace(-1.2, 1.2, 100)

x, y = np.meshgrid(wx, wy)

z = x**2 + y**2

contourplot = plt.contour(x, y , z, np.arange(0.1, 1.1, 0.1), cmap='Paired')

plt.axis('equal')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Contours')

plt.clabel(contourplot)
```

```python
plt.figure(figsize=(7,7))

wyg = np.linspace(-1.2, 1.3, 100)
wxg = np.linspace(-1.2, 1.3, 100)
x, y = np.meshgrid(wxg, wyg)

z = x**2 + y**2

ax = plt.axes()

ax.contour(x, y, z, np.arange(0.1, 1.1, 0.1), cmap = 'Paired')

ax.set_aspect('equal')

wyg1 = np.arange(-1.2, 1.3, 0.18)
wxg1 = np.arange(-1.2, 1.3, 0.24)

x1, y1 = np.meshgrid(wxg1, wyg1)

gradhx = 2 * x1

gradhy = 2 * y1

ax.quiver(x1, y1, gradhx, gradhy, color='m', width = 0.0015)

ax.set_aspect('equal')
ax.set_title('Quivers and Contour Plot using analytical grad(h)')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
```

### Example 2

Plot a contour and gradients of the function $ z = e^{(-x^2 - y^2)} + 0.5 e^{(-(x-1.5)^2 - 2y^2)} $

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 7))

wyg = np.linspace(-2,2,100)

wxg = np.linspace(-2,3,100)

x, y = np.meshgrid(wxg,wyg)

z = np.exp(-x**2 - y**2) + 0.5 * np.exp(-(x-1.5)**2 - 2*(y**2))

contourplot = plt.contour(x, y , z, np.arange(0.1,1.1,0.1), cmap='Paired')

plt.axis('equal')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Contours')

plt.clabel(contourplot)
```

```python
plt.figure(figsize=(7, 7))

wyg = np.linspace(-2, 2, 100)
wxg = np.linspace(-2, 3, 100)
x, y = np.meshgrid(wxg, wyg)

z = np.exp(-x**2 - y**2) + 0.5 * np.exp(-(x-1.5)**2 - 2 * (y**2))

ax = plt.axes()

ax.contour(x, y, z, np.arange(0.1, 1.1, 0.1), cmap = 'Paired')

ax.set_aspect('equal')

wyg1 = np.arange(-2, 2, 0.18)
wxg1 = np.arange(-2, 3, 0.24)

xg1, yg1=np.meshgrid(wxg1, wyg1)

gradhx = -2 * xg1 * np.exp(-xg1**2 - yg1**2) - (np.exp(-((xg1-1.5)**2) - 2 * yg1**2)) * (xg1 - 1.5)

gradhy = -2 * yg1 * np.exp(-xg1**2 - yg1**2) - (np.exp(-((xg1-1.5)**2) - 2 * yg1**2)) * 2 * yg1

ax.quiver(xg1, yg1, gradhx, gradhy, color='m', width=0.0015)

ax.set_aspect('equal')
ax.set_title('Quivers and Contour Plot using analytical grad(h)')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
```
